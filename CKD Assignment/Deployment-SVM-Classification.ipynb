```python
import pickle
import pandas as pd

# Load model and scaler
loaded_model  = pickle.load(open("finalised_model_SVM.sav", 'rb'))
loaded_scaler = pickle.load(open("scaler_SVM.sav", 'rb'))

# patient values

patient = [[48, 80, 4, 0, 200, 100, 7.2, 130, 5.5, 8.5, 28, 11000, 3.2,
             0,  1,  0, 0,  0,  0,   1,  0,   1,  1,  0,    0,  1,  1]]

# Scale and predict
patient_scaled = loaded_scaler.transform(patient)
result         = loaded_model.predict(patient_scaled)

print(result)
if result[0]:
    print("CKD DETECTED")
else:
    print("NO CKD DETECTED")
```

```python
#enter Values
age   = float(input("Age: "))
bp    = float(input("Blood Pressure: "))
al    = float(input("Albumin (0-5): "))
su    = float(input("Sugar (0-5): "))
bgr   = float(input("Blood Glucose: "))
bu    = float(input("Blood Urea: "))
sc_val= float(input("Serum Creatinine: "))
sod   = float(input("Sodium: "))
pot   = float(input("Potassium: "))
hrmo  = float(input("Hemoglobin: "))
pcv   = float(input("Packed Cell Volume: "))
wc    = float(input("White Blood Cell Count: "))
rc    = float(input("Red Blood Cell Count: "))
print("\n Enter 1 for Yes, 0 for No\n")
sg_b   = int(input("Specific Gravity 1.010: "))
sg_c   = int(input("Specific Gravity 1.015: "))
sg_d   = int(input("Specific Gravity 1.020: "))
sg_e   = int(input("Specific Gravity 1.025: "))
rbc    = int(input("Red Blood Cells Normal: "))
pc     = int(input("Pus Cells Normal: "))
pcc    = int(input("Pus Cell Clumps Present: "))
ba     = int(input("Bacteria Present: "))
htn    = int(input("Hypertension: "))
dm     = int(input("Diabetes: "))
cad    = int(input("Coronary Artery Disease: "))
appet  = int(input("Good Appetite: "))
pe     = int(input("Pedal Edema: "))
ane    = int(input("Anemia: "))
```

```python
 #Build DataFrame with column names
new_patient = pd.DataFrame([[
    age, bp, al, su, bgr, bu, sc_val, sod, pot, hrmo, pcv, wc, rc,
    sg_b, sg_c, sg_d, sg_e, rbc, pc, pcc, ba, htn, dm, cad, appet, pe, ane
]], columns=[
    'age','bp','al','su','bgr','bu','sc','sod','pot','hrmo','pcv','wc','rc',
    'sg_b','sg_c','sg_d','sg_e','rbc_normal','pc_normal',
    'pcc_present','ba_present','htn_yes','dm_yes','cad_yes',
    'appet_yes','pe_yes','ane_yes'
])
```

```python
new_patient_scaled = loaded_scaler.transform(new_patient)
```

```python
result      = loaded_model.predict(new_patient_scaled)
probability = loaded_model.predict_proba(new_patient_scaled)
```

```python
print("\n--- RESULT ---")
if result[0]:
    print("CKD DETECTED")
else:
    print("NO CKD DETECTED")
print(f"CKD Chance    : {probability[0][1]*100:.2f}%")
print(f"No CKD Chance : {probability[0][0]*100:.2f}%")
```
